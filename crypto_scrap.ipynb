```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
from datetime import datetime

# Initialize the CoinGecko API client
cg = CoinGeckoAPI()

# Fetch the data
data = cg.get_coins_markets(vs_currency='inr')
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# Add a timestamp column
df['Timestamp'] = datetime.now()
df
```

```python
columns = ['id', 'symbol', 'name', 'current_price', 'market_cap', 'total_volume', 'high_24h', 'low_24h', 'price_change_percentage_24h', 'Timestamp']
df = df[columns]

# Display the data
print(df)
```

```python
# Function to analyze and display daily trends
def analyze_daily_trends(df):
    # Convert the Timestamp column to datetime
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])

    # Set the Timestamp as the index
    df.set_index('Timestamp', inplace=True)

    # Group by date
    daily_data = df.groupby(df.index.date).agg({
        'current_price': ['mean', 'max', 'min'],
        'market_cap': 'sum',
        'total_volume': 'sum',
        'price_change_percentage_24h': 'mean'
    })

    # Display the daily trends
    print("\nDaily Trends:")
    print(daily_data)

# Analyze and display the daily trends
analyze_daily_trends(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Top 10 Cryptocurrencies by Market Cap
top_10_market_cap = df.nlargest(10, 'market_cap')
plt.figure(figsize=(14, 7))
sns.barplot(x='market_cap', y='name', data=top_10_market_cap, palette='viridis')
plt.title('Top 10 Cryptocurrencies by Market Cap')
plt.xlabel('Market Cap (USD)')
plt.ylabel('Cryptocurrency')
plt.show()
```

```python
# Price Distribution
plt.figure(figsize=(14, 7))
sns.histplot(df['current_price'], bins=50, kde=True)
plt.title('Cryptocurrency Price Distribution')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Price Change Percentage vs Current Price
plt.figure(figsize=(14, 7))
sns.scatterplot(x='current_price', y='price_change_percentage_24h', data=df, hue='name', legend=False, palette='tab10')
plt.title('Price Change Percentage (24h) vs Current Price')
plt.xlabel('Current Price (USD)')
plt.ylabel('Price Change Percentage (24h)')
plt.xscale('log')
plt.show()
```
